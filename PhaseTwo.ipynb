###Handling Missing Values:


```python
import pandas as pd

data = pd.read_csv('StudentPerformanceFactors.csv')
mode_value_teacher_quality = data['Teacher_Quality'].mode()[0]
data['Teacher_Quality'] = data['Teacher_Quality'].fillna(value=mode_value_teacher_quality)

mode_value_parental_education = data['Parental_Education_Level'].mode()[0]
data['Parental_Education_Level'] = data['Parental_Education_Level'].fillna(value=mode_value_parental_education)

mode_value_distance_from_home = data['Distance_from_Home'].mode()[0]
data['Distance_from_Home'] = data['Distance_from_Home'].fillna(value=mode_value_distance_from_home)

missing_values = data.isnull().sum()
print('Missing values per column:')
print(missing_values)

data.to_csv('Cleaned_dataset.csv', index=False)
```

1.  As we mentioned earlier, our dataset contains three columns with missing values.

2. we handle these missing values by calculating the mean value for the  column Distance_from_Home amd the mode value to the columns Parental_Education_Level and Distance_from_Home, and then we replace the missing values with the resulting mean and mode.

- We chose mean imputation for numeric columns because:
 - It preserves the data distribution and avoids introducing extreme values.
 - It's simple, effective, and works well in large datasets with minimal missing data.

- We chose mode imputation for categorical columns because:
 - It keeps the most frequent category, ensuring data consistency.
 - It prevents data loss by filling missing values without dropping rows.


### Handling outliers:
Before handling the outliers, we selected the IQR (Interquartile Range) method based on our dataset to determine how many outliers were present.



```python

import numpy as np
outlier_threshold = 1.5

def count_outliers(column_data):
    q1 = np.percentile(column_data, 25)
    q3 = np.percentile(column_data, 75)
    iqr = q3 - q1
    upper_bound = q3 + outlier_threshold * iqr
    lower_bound = q1 - outlier_threshold * iqr
    outliers = (column_data > upper_bound) | (column_data < lower_bound)
    return sum(outliers)

numeric_columns = df.select_dtypes(include=[np.number]).columns

outlier_counts = {}
total_rows_with_outliers = 0

for column in numeric_columns:
    outliers = count_outliers(df[column])
    outlier_counts[column] = outliers
    total_rows_with_outliers += outliers

total_rows = len(df)

print("Outlier Counts:")
for column, count in outlier_counts.items():
    print(f"{column}: {count} rows with outliers")

print(f"Total Rows with Outliers: {total_rows_with_outliers}")
```

Attendance: Interestingly, there are no outliers in attendance. This indicates that students generally have stable attendance rates, which is a good sign.

Sleep_Hours: Just like attendance, there are no outliers for sleep hours. It seems that students have fairly consistent sleep patterns.

Previous_Scores: Again, no outliers were found in this variable. It looks like students past scores are relatively similar.

Physical_Activity: No outliers here either, indicating that students tend to have a consistent level of physical activity.

```python
import pandas as pd

data2 = pd.read_csv('Cleaned_dataset.csv')

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

columns_to_check = ['Hours_Studied', 'Tutoring_Sessions', 'Exam_Score']

for column in columns_to_check:
    data2 = remove_outliers_iqr(data2, column)

data2.to_csv('Cleaned_dataset.csv', index=False)
```

After handling outliers, we focuses on improving data quality by automatically removing extreme values from selected columns using the Interquartile Range IQR method. The IQR technique is especially effective for dealing with skewed data, as it eliminates outliers that could distort the results of analysis or machine learning models. By filtering out these outliers, we ensure that only the most relevant and reliable data points remain. Once the dataset is cleaned, it is saved as a CSV file, making it ready for further analysis or model training. This process improves the overall accuracy and consistency of the dataset.

# Data Transmation:

### Encoding:


   The primary goal of using LabelEncoder is to convert categorical data into a numeric format that can be easily processed by machine learning algorithms. Categorical features often represent qualitative information that needs to be quantitatively analyzed.
   

In the encoding method, we represent the categorical variables as follows:
- Gender: Original categories like 'Male' and 'Female' are transformed into numeric values, where 'Male' might 1 and 'Female' become 0.


- Distance_from_Home:This column contains categorical representations of distance, such as 'Far', 'Near', or 'Moderate'. These categories are converted into integers, where 'Far' might become 2, 'Near' 0, and 'Moderate' 1.

- Parental_Education_Level:Represents the education levels of parents, such as 'High School' or 'Bachelor's Degree'. Each level is assigned a numeric value, with 'High School' becoming 0 and 'Bachelor's Degree' becoming 1.

- Learning_Disabilities:Contains binary categories like 'Yes' or 'No', which are transformed into 0 and 1.

- Peer_Influence:Categories that describe the influence of peers, such as 'High', 'Low', or 'Moderate', are replaced by numeric values, where 'High' become 2, 'Low' 0, and 'Moderate' 1.

- School_Type:Different school types, such as 'Private' and 'Public', are encoded as integers, where 'Private' might become 1 and 'Public' 0.

- Teacher_Quality:Reflects categorical assessments of teacher quality, such as 'Excellent', 'Good', or 'Poor', each mapped to a number, with 'Excellent' becoming 2, 'Good' 1, and 'Poor' 0.

- Family_Income:Represents income categories, such as 'Low', 'Medium', or 'High', and each is transformed into a corresponding number, where 'Low' become 0, 'Medium' 1, and 'High' 2.

- Internet_Access:Refers to whether the student has access to the internet, typically a binary value where 'Yes'become 1 and 'No' 0.

- Motivation_Level:Represents the student's motivation levels, such as 'High', 'Moderate', or 'Low', each mapped to a numeric label, where 'High' becomes 2, 'Moderate' 1, and 'Low' 0.

-  Extracurricular_Activities:Encodes whether a student participates in extracurricular activities, where 'Yes' become 1 and 'No' 0.

- Access_to_Resources: Indicates the student's access to educational resources, such as 'High', 'Low', or 'Moderate', which are replaced by integers, with 'High' becoming 2, 'Low' 0, and 'Moderate' 1.

- Parental_Involvement:Reflects how involved parents are in the student's education, such as 'High', 'Low', or 'Moderate', each involvement level mapped to a numeric value, where 'High' becomes 2, 'Low' 0, and 'Moderate' 1.

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

data2 = pd.read_csv('Cleaned_dataset.csv')
le = LabelEncoder()
categorical_columns = ['Gender', 'Distance_from_Home','Parental_Education_Level','Learning_Disabilities','Peer_Influence','School_Type',
                       'Teacher_Quality','Family_Income','Internet_Access','Motivation_Level','Extracurricular_Activities',
                       'Access_to_Resources','Parental_Involvement']

for col in categorical_columns:
    data2[col] = le.fit_transform(data2[col])

print(data2)
data2.to_csv('encoding.csv', index=False)
```


###Normaliztion:

```python
import pandas as pd

data1 = pd.read_csv('encoding.csv')
data1 = pd.DataFrame(data1)

columns_to_normalize = ['Hours_Studied', 'Attendance', 'Sleep_Hours',
                        'Previous_Scores', 'Tutoring_Sessions',
                        'Physical_Activity', 'Exam_Score','Tutoring_Sessions']

# Decimal scaling normalization
for column in columns_to_normalize:
    max_abs_value = data1[column].abs().max()
    data1[column] = data1[column] / (10 ** len(str(int(max_abs_value))))

print("DataFrame after Decimal Scaling Normalization:")
print(data1)
data1.to_csv('normalization.csv', index=False)
```

We chose decimal scaling normalization because it adjusts values to a comparable scale without changing their distribution. It's simple, efficient, and ideal for handling varying ranges in our dataset (e.g., Hours_Studied, Attendance, Exam_Score), ensuring easy comparison across columns while preserving the relationships between data points.



### Discretization categorizes exam scores into three labels('Low' 'Mid' and 'High'):

```python
import pandas as pd
data1 = pd.read_csv('normalization.csv')
income_labels = ['Low', 'Mid', 'High']
data1['Exam_Score'] = pd.qcut(data1['Exam_Score'], q=3, labels=income_labels)
print(data1)
data1.to_csv('after_discretization.csv', index=False)

```

In this step, we converted the Exam Score variable into categorical labels (Low, Mid, and High).
 This transformation is important for classification tasks as it allows the use of algorithms that require a categorical target variable. Discretizing the exam scores helps improve model performance, enhances interpretability for stakeholders, and addresses potential class imbalance, making our classification models more robust.

#Feature selection




Feature selection is an essential step in the process of preparing data for machine learning models. It assists in pinpointing the most important features that play a role in predicting the target variable. This can enhance model performance, minimize overfitting, and reduce training times.

### Filter Method: SelectKBest with ANOVA F-test

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif
df = pd.read_csv('after_discretization.csv')
le = LabelEncoder()
X = df.drop(columns=['Exam_Score'])
y = df['Exam_Score']
X = X.loc[:, (X != X.iloc[0]).any()]
X = X.select_dtypes(include=['int64', 'float64'])
selector = SelectKBest(score_func=f_classif, k=5)
X_new = selector.fit_transform(X, y)

# Display the selected features
selected_features = X.columns[selector.get_support()]
print("Selected Features:", selected_features)
```

### Wrapper Method:RFE

```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('after_discretization.csv')
X = data.drop(columns=['Exam_Score'])
y = data['Exam_Score']

label_encoders = {}
for column in X.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X[column] = le.fit_transform(X[column])
    label_encoders[column] = le

X_encoded = pd.get_dummies(X, drop_first=True)

model = LogisticRegression(max_iter=1000)

rfe = RFE(model, n_features_to_select=5)
rfe_fit = rfe.fit(X_encoded, y)

selected_features_rfe = X_encoded.columns[rfe.support_]
print("Selected features using RFE:", selected_features_rfe.tolist())
```

### Embedded Method:Lasso

```python
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('after_discretization.csv')

X = data.drop(columns=['Exam_Score'])
y = data['Exam_Score']

label_encoders = {}
for column in X.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X[column] = le.fit_transform(X[column])
    label_encoders[column] = le

y = LabelEncoder().fit_transform(y)  # Encode target variable

X_encoded = pd.get_dummies(X, drop_first=True)

lasso = Lasso(alpha=0.01)
lasso.fit(X_encoded, y)

selected_features_lasso = X_encoded.columns[lasso.coef_ != 0]
print("Selected features using Lasso:", selected_features_lasso.tolist())
```

### Feature Selection Steps

1. **SelectKBest Method**:
   - We applied the **SelectKBest** method using ANOVA F-value to identify significant features.
   - The features selected included **Hours Studied**, **Attendance**, **Previous Scores**, **Tutoring Sessions**, and **Peer Influence**.

2. **Recursive Feature Elimination (RFE)**:
   - We utilized **RFE** with a Logistic Regression model to further assess feature importance.
   - This method highlighted **Hours Studied**, **Attendance**, **Previous Scores**, **Tutoring Sessions**, and **Learning Disabilities** as important predictors.

3. **Lasso (L1 Regularization)**:
   - We implemented **Lasso** for feature selection, resulting in a broader selection of relevant features.
   - The selected features included **Parental Involvement**, **Access to Resources**, **Family Income**, **Teacher Quality**, **Peer Influence**, **Parental Education Level**, and **Distance from Home**.

4. **Final Selection**:
   - We chose **Lasso** as our primary feature selection method due to its effectiveness in shrinking the coefficients of less significant features to zero, simplifying the model while retaining key predictors.
   - Lasso also helps manage multicollinearity among features, enhancing interpretability and robustness in our predictions.

By following these steps, we ensured a thorough evaluation of our dataset, selecting the most relevant features to improve model performance and reduce overfitting.
